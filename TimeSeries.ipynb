```python
from tslearn.clustering import TimeSeriesKMeans
import numpy as np
```

```python
w = TimeSeriesKMeans(n_clusters=3, metric='dtw')
```

```python
import pandas as pd
dfL = []
for i in range(1, 13):
  if i >= 10:
    n = str(i)
  else:
    n = '0' + str(i)

  df = pd.read_csv('/content/drive/MyDrive/research/ACTIVITY_Sergei/Data/Demand/2020-' + n + '-PD.csv')
  dfL.append(list(df['value']))
```

```python
w.fit(dfL[:, :100, :])
```
